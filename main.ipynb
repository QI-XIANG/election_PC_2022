```python
import pandas as pd
df = pd.read_csv("盧秀燕市長選舉資料\incomes.csv")
df
```

```python
df[df["收支科目"]=="營利事業捐贈收入"].sort_values(by='收入金額',ascending=False)[0:5]
```

```python
def getData_MS(df):
    keys = list(df.groupby(["收支科目"]).size().keys())
    size_group = [df.groupby(["收支科目"]).size()[i] for i in keys]
    return keys, size_group

def DataToDataFrame_MS(df):
    keys, size_group = getData_MS(df)
    cleanData = pd.DataFrame({"收支科目":keys, "資料筆數":size_group})
    return cleanData

DataToDataFrame_MS(df)
```

```python
def getData_MA(df):
    keys = list(df.groupby(["收支科目"]).size().keys())
    total_money = [int(df[df["收支科目"]==i]["收入金額"].sum()) for i in keys]
    return keys, total_money

def DataToDataFrame_MA(df):
    keys, total_money = getData_MA(df)
    cleanData = pd.DataFrame({"收支科目":keys, "總金額":total_money})
    return cleanData

DataToDataFrame_MA(df)
```

```python
def findLoc(df):
    loc = str(df["地址"])[0:3]
    if(loc == "台中市" or loc == "大里區"):
        loc = "臺中市"
    if(loc == "桃園縣"):
        loc = "桃園市"
    if(loc == "台北市"):
        loc = "臺北市"
    if(loc == "nan"):
        loc = "無資料"
    return loc
df["縣市"] = df.apply(findLoc,axis=1)
```

```python
df["地址"]
```

```python
def getData_LA(df):
    keys = list(df.groupby(["縣市"]).size().keys())
    total_money = [df.groupby(["縣市"]).size()[i] for i in keys]
    return keys, total_money

def DataToDataFrame_LA(df):
    keys, total_money = getData_LA(df)
    cleanData = pd.DataFrame({"縣市":keys, "捐贈筆數":total_money}).sort_values(["捐贈筆數"],ascending=False,ignore_index=True)
    cleanData = cleanData.drop(cleanData[cleanData["縣市"]=="無資料"].index,axis=0)
    return cleanData

locData = DataToDataFrame_LA(df)
sum = locData[4:]["捐贈筆數"].sum()
locData = locData.append({"縣市":"其他縣市","捐贈筆數":sum},ignore_index=True)
locData = locData.drop([i for i in range(4,len(locData)-1,1)],axis=0)
locData = locData.reset_index().drop("index",axis=1)
locData
```

```python
Advanced_data = pd.concat([DataToDataFrame_MS(df),DataToDataFrame_MA(df),locData],axis=1).T.drop_duplicates().T
Advanced_data.to_csv("./cleanDataSet/total_money_type_3.csv",index=False)
```

```python
def formatDollar(df):
    dollar = "${:0,.0f}".format(df['收入金額'])
    return dollar

df['收入金額'] = df.apply(formatDollar,axis=1)
```
